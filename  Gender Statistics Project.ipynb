```
import itertools
# Finding the right combination of p,d,q (keep the lowest AIC value)

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

def combination (dataframe_column):
  for param in pdq:
    for param_seasonal in seasonal_pdq:
      try:
        mod = sm.tsa.statespace.SARIMAX(dataframe_column,
                        order=param,
                        seasonal_order=param_seasonal,
                        enforce_stationarity=False,
                        enforce_invertibility=False)
        results = mod.fit()

        print('SARIMAX{}x{} - AIC:{}'.format(param, param_seasonal, results.aic))
      except:
        continue
        
combination(ArabWorld['School enrollment, primary, female (% gross)'])
```
