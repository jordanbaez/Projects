```
derma_cat = derma.drop('Age', axis =1)

derma_dummies = pd.concat([pd.get_dummies(derma_cat[col], drop_first=True) 
                           for col in derma_cat], axis=1, keys=derma_cat.columns)

derma_dummies.shape
```

```
derma_dummies.head()
```

```
from sklearn.model_selection import train_test_split

X = derma_dummies.drop('erythema',axis=1) 
y = derma.erythema
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

# here, traiing set first and what we want to predict second
```

```
from sklearn import linear_model
model = linear_model.LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')

results=model.fit(X_train, y_train)
print('Model score :',results.score(X_train, y_train))
```

```
from sklearn.metrics import accuracy_score

x_pred = results.predict(X_train)

#print('score train set',accuracy_score( x_pred,y_train))

y_pred = results.predict(X_test)

print('Accuracy score test set :',accuracy_score( y_pred,y_test))
```

```
from sklearn.metrics import balanced_accuracy_score
print('Test set balanced accuracy score' , balanced_accuracy_score(y_pred,y_test))
```

```
from sklearn import metrics

#cnf_matrix_test = metrics.confusion_matrix(y_test, y_pred)
#cnf_matrix_train = metrics.confusion_matrix(y_train, y_pred_train)
#cnf_matrix_train

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred,average='micro'))
print("Recall:",metrics.recall_score(y_test, y_pred,average='micro'))
```

```
from sklearn.metrics import precision_score, recall_score, f1_score
print('train precision' , precision_score(x_pred,y_train, average = 'micro'),
'recall' , recall_score(x_pred,y_train, average = 'micro'),
'f1_score' , f1_score(x_pred,y_train, average = 'micro'))

print('test precision' , precision_score(y_pred,y_test, average = 'micro'),
'recall' , recall_score(y_pred,y_test, average = 'micro'),
'f1_score' , f1_score(y_test,y_pred, average = 'micro'))
```
